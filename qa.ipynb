```python
# Check the shape and columns
print("DataFrame shape:", df.shape)
print("\nColumns:", df.columns.tolist())
print("\nFirst few rows:")
df.head()
```

```python
# Check if there's a 'category' column and see the value counts
if 'category' in df.columns:
    print("Category counts:")
    print(df['category'].value_counts())
else:
    # List columns to help identify the category column
    print("Available columns:", df.columns.tolist())
```

```python
# Sample 50 rows randomly for each category
# Replace 'category' with the actual column name if different
sampled_df = df.groupby('category', group_keys=False).apply(
    lambda x: x.sample(n=min(50, len(x)), random_state=42)
).reset_index(drop=True)

print(f"Original dataframe shape: {df.shape}")
print(f"Sampled dataframe shape: {sampled_df.shape}")
print("\nSamples per category:")
print(sampled_df['category'].value_counts())
```
